### Process 'Datetime' column

```python
new = data['ReadingDateTime'].str.split(" ", n=1, expand=True)
```

```python
data['Date']=new[0]
```

```python
data['Time']=new[1]
```

```python
data.drop(columns=['ReadingDateTime'], inplace=True)
```

```python
data
```

```python
from datetime import datetime
```

```python
data['Date'] = data['Date'].apply(lambda x:datetime.strptime(x, '%d/%m/%Y'))
```

```python
data.info()
```

```python
data['Time']=data['Time'].apply(lambda x:datetime.datetime.strptime(x, '%H:%M').time())
```
